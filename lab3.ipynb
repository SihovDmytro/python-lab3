```python
import pandas as pd
import numpy as np

def maxRenewable(dataframe):
    return dataframe.sort_values(by=["% Renewable"]).tail(1)['% Renewable']

def avgGDP(dataframe):
    return dataframe.iloc[:, 11:21].mean(axis=1).sort_values(ascending=False)

#1 Загрузити файл з даними про споживання енергії “En_In.xls”,
columns = ['Country','Energy Supply','Energy Supply per Capita','% Renewable']
df1 = pd.read_excel('En_In.xls', sheet_name='Energy', usecols="C:F", skiprows=17, nrows=227, names=columns)
# df1

#2 Переведіть дані із стовпчику ‘Energy Supply’ в гДж (Примітка: в 1ПДж = 1000000 ГДж).
# Для всіх країн, у яких відсутні дані (наприклад, дані з"....."), переконайтеся, що це відображається як значення np.NaN.

df1['Energy Supply'] = pd.to_numeric(df1['Energy Supply'], errors='coerce')* 1000000
df1.loc[df1['Energy Supply per Capita'] == '...', 'Energy Supply per Capita'] = np.NaN
# df1

# 3. Перейменуйте наступний список країн:
# "Republic of Korea": "South Korea",
# "United States of America": "United States",
# "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
# "China, Hong Kong Special Administrative Region": "Hong Kong"

df1['Country'] = df1['Country'].replace('Republic of Korea', 'South Korea', regex=True)
df1['Country'] = df1['Country'].replace('United States of America', 'United States', regex=True)
df1['Country'] = df1['Country'].replace('United Kingdom of Great Britain and Northern Ireland', 'United Kingdom', regex=True)
df1['Country'] = df1['Country'].replace('China, Hong Kong Special Administrative Region', 'Hong Kong', regex=True)
# df1['Country'].values.tolist()

# 4. Існує також декілька країн, що мають в назві цифри та/або дужки.
# Обов’язково видаліть їх, напр. 'Bolivia (Plurinational State of)'
# повинна бути ‘Bolivia’, ‘Switzerland17’ повинна бути Switzerland’.

df1['Country'] = df1['Country'].replace('\d+', '', regex=True)
df1['Country'] = df1['Country'].replace('\([^)]*\)', '', regex=True)
df1['Country'] = df1['Country'].str.strip()
df1.loc[df1['Country'].isin(['American Samoa', 'South Korea', 'Bolivia'])]

# 5. Далі, завантажте дані про ВВП з файлу „gpd.csv”, що містить дані
# Світового банку про ВВП країн від 1960 до 2015 року.
# Обов’язково не включайте заголовок до датафрейму і перейменуйте
# наступний список країн:
# "Korea, Rep.": "South Korea",
# "Iran, Islamic Rep.": "Iran",
# "Hong Kong SAR, China": "Hong Kong"
df2 = pd.read_csv('gpd.csv', skiprows=4)
df2 = df2.rename(columns={df2.columns[0]: 'Country'}) 

df2['Country'] = df2['Country'].replace('Korea, Rep.', 'South Korea', regex=True)
df2['Country'] = df2['Country'].replace('Iran, Islamic Rep.', 'Iran', regex=True)
df2['Country'] = df2['Country'].replace('Hong Kong SAR, China', 'Hong Kong', regex=True)
# df2

# 6. Завантажте дані з файлу „scimagojr.xlsx”, який класифікує країни на основі
# їхніх публікацій в журналах у галузі енергетичного машинобудування та
# енергетичних технологій.
df3 = pd.read_excel('scimagojr.xlsx', sheet_name='Sheet1')
# df3

# Приєднайте три набори даних із завдань 1-6 до нового набору даних
# (використовуючи перетин назв країн).

df_merged = pd.merge(df3, pd.merge(df1, df2)).set_index('Country')
df_merged = df_merged.sort_values(by=['Rank']).head(15).drop(df_merged.iloc[:, 10:59], axis=1)
pd.set_option('float_format', '{:f}'.format)
# df_merged

# 8. Створіть функцію, щоб визначити середній ВВП країн за останні 10 років.

# avgGDP(df_merged)


# 10. Створіть функцію, щоб визначити, в якій країні встановлено максимум
# поновлюваних джерел енергії (% Renewable) та який саме відсоток.

tuple(maxRenewable(df_merged).items())
```
